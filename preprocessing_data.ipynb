# Hall of Fame Stats

```python
# read in Hall of Fame data
hof_df = pd.read_csv(f'{file_dir}HallOfFame.csv')
hof_df.head()
```

```python
# check how many unique playerID's in the original csv
hof_df['playerID'].nunique()
```

```python
# sort inducted yes & no, then drop duplicate no's to get 1 row per playerID
# hof_df['inducted'] = pd.Categorical(hof_df['inducted'], categories=['Y','N'], ordered=True)
# hof_df = hof_df.sort_values(['playerID','inducted'], ascending=[True,False]).drop_duplicates('playerID')


# drop duplicate playerID's, but keep last row
# if player inducted, last instance of playerID will have Y
# if player not inducted, last instance will be N
hof_df = hof_df.drop_duplicates(subset='playerID',keep='last')
hof_df = hof_df.set_index('playerID')
hof_df.head()
```

```python
# confirm correct number of rows
hof_df.shape
```

```python
# write to new csv file
hof_df.to_csv(f'{file_dir}HOF.csv')
```

# Batting Stats

```python
# read in Batting stats csv
batting_df = pd.read_csv(f'{file_dir}Batting.csv')
batting_df.head()
```

```python
# combine all stats for each playerID to get career stats
batting_df = batting_df.groupby(['playerID']).sum()
batting_df.drop(['yearID','stint'],axis=1, inplace=True)
batting_df.tail()
```

```python
# compute batting average
batting_df['AVG'] = batting_df['H'] / batting_df['AB'] 
batting_df.tail()
```

```python
# write to new csv file
batting_df.to_csv(f'{file_dir}Batting_cleaned.csv')
```

# Fielding Stats

```python
# read in Fielding stats csv
fielding_df = pd.read_csv(f'{file_dir}Fielding.csv')
fielding_df.head()
```

```python
# combine all stats for each playerID to get career stats
fielding_df = fielding_df.groupby(['playerID']).sum()
fielding_df.drop(['yearID','stint'],axis=1, inplace=True)
fielding_df.tail()
```

```python
# write to new csv file
fielding_df.to_csv(f'{file_dir}Fielding_cleaned.csv')
```

# Pitching Stats

```python
# read in Pitching stats csv
pitching_df = pd.read_csv(f'{file_dir}Pitching.csv')
pitching_df.head()
```

```python
# check how many unique playerID's in the original csv
pitching_df['playerID'].nunique()
```

```python
# combine all stats for each playerID to get career stats
# need dictionary to be sure the OppBA & ERA are means, 
# the remaining colums are all sums
pitching_df = pitching_df.groupby('playerID').agg({'W':'sum','L':'sum','G':'sum','GS':'sum','SHO':'sum','SV':'sum','IPouts':'sum','H':'sum','ER':'sum','HR':'sum','BB':'sum','SO':'sum','BAOpp':'mean','ERA':'mean','IBB':'sum','WP':'sum','HBP':'sum','BK':'sum','BFP':'sum','GF':'sum','R':'sum','SH':'sum','SF':'sum','GIDP':'sum'})
pitching_df.tail()
```

```python
# double check count of unique playerID
pitching_df.shape
```

```python
# write to new csv file
pitching_df.to_csv(f'{file_dir}Pitching_cleaned.csv')
```

# Awards Stats

```python
# read in Awards stats csv
awards_df = pd.read_csv(f'{file_dir}AwardsPlayers.csv')
awards_df.head()
```

```python
# get count of awards per playerID
awards_df = awards_df.groupby(['playerID']).count()
awards_df.tail()
```

```python
awards_df.columns = ['award_count','year_count','lgID_count','tie_count','notes_count']
awards_df.head()
```
