```python
import pandas as pd
import numpy as np
```

```python
file_dir = "C:/Users/nickl/Documents/DataViz/MyRepo/02-Assignments/Project_Finale/data/"
```

```python
hof_df = pd.read_csv(f'{file_dir}HallOfFame.csv')
hof_df.head()
```

```python
# sort inducted yes & no, then drop duplicate no's to get 1 row per playerID
hof_df['inducted'] = pd.Categorical(hof_df['inducted'], categories=['Y','N'], ordered=True)
hof_df = hof_df.sort_values(['playerID','inducted'], ascending=[True,False]).drop_duplicates('playerID')
hof_df.head()
```

```python
hof_df.to_csv(f'{file_dir}HOF.csv')
```

```python
batting_df = pd.read_csv(f'{file_dir}Batting.csv')
batting_df.head()
```

```python
batting_df = batting_df.groupby(['playerID']).sum()
batting_df.tail()
```

```python
batting_df['AVG'] = batting_df['H'] / batting_df['AB'] 
batting_df.tail()
```

```python
batting_df.to_csv(f'{file_dir}Batting_cleaned.csv')
```

```python
fielding_df = pd.read_csv(f'{file_dir}Fielding.csv')
fielding_df.head()
```

```python
fielding_df = fielding_df.groupby(['playerID']).sum()
fielding_df.tail()
```

```python
fielding_df.to_csv(f'{file_dir}Fielding_cleaned.csv')
```
